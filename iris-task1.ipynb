# IMPORTING ALL THE NECESSARY LIBRARIES 

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
```

# LOADING THE *IRIS* DATASET

```python
iris = load_iris()
```

```python
iris
```

## Studying the Iris dataset and making a dataframe accordingly

```python
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['Species'] = iris.target_names[iris.target]
```

```python
df.head()
```

```python
df.isnull().sum()
```

### To use a machine learning algorithm we need all the values in a integer format but in our dataframe the column 'Species' contains string values so to deal with that we use LabelEncoder from sklearn

```python
le = LabelEncoder()
```

```python
df['Species'] = le.fit_transform(df['Species'])
df.sample(10)
```

## Now our data is ready to be trained 
### We split the data in 2 parts - Train and Test
> **The training set is a portion of your dataset that you use to train your machine learning model. It's the part of the data that the model learns from.**

> **The test set is a separate portion of your dataset that you set aside for evaluating the performance of your trained machine learning model.**

```python
#train - 90
#test - 10
X = df.drop(columns=['Species'])
Y = df['Species']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.10)
```

### We splitted the data in the ratio of 9:1 
#### (Depending on the model the ratio could change)

```python
# logistic regression
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

## We will be using LogisticRegression model as it works best when classes in the dataset are linearly seperable
### (as they are to some extent in *IRIS* dataset)

### Now we are training the model by fitting in the dataset

```python
model.fit(x_train, y_train)
```

### Now we can check the accuracy on our trained model

```python
model.score(x_test, y_test)
```

## WE GOT AN ACCURACY OF 100%
