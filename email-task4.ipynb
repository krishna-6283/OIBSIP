# EMAIL SPAM DETECTION

```python
import pandas as pd
import numpy as np
```

## IMPORTING THE DATASET

```python
df = pd.read_csv('/kaggle/input/sms-spam-collection-dataset/spam.csv', encoding='ISO-8859-1')
df
```

## Dropping irrelevant columns

```python
df = df.drop(columns=['Unnamed: 2','Unnamed: 3','Unnamed: 4'])
```

```python
df
```

## Describing the dataframe on the basis of *v1* column

```python
df.groupby('v1').describe()
```

## As v1 contains string values
## it is needed to be converted to numerical values
> **'ham' = 0**

> **'spam' = 1**

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['v1'] = le.fit_transform(df['v1'])
```

```python
df
```

## Splitting the dataset into train and test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.v2,df.v1)
```

## Using count vectorizer feature to transfrom text data into numeric values

```python
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
X_train_count = v.fit_transform(X_train.values)
X_train_count.toarray()[:2]
X_test_count = v.transform(X_test)
```

## Using the Naive Bayees Algorithm

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_count,y_train)
```

```python
model.score(X_test_count, y_test)
```

# We can also use the *PIPELINE* feature from sklearn to do the same thing

```python
from sklearn.pipeline import Pipeline
clf = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb', MultinomialNB())
])
```

```python
clf.fit(X_train, y_train)
```

```python
clf.score(X_test,y_test)
```

# We got an accuracy of approximately 99%
